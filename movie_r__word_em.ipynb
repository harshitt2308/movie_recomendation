```python
credits_column_renamed = credits.rename(index=str, columns={"movie_id": "id"})
```

```python
movies = movies_df.merge(credits_column_renamed,on ='id')
```

```python
movies.head()
```

```python
movies = movies[['id','title_x','overview','genres','keywords','cast','crew']]
```

```python
movies.head(5)
```

```python
import ast
```

```python
def convert(text):
  L = []
  for i in ast.literal_eval(text):
    L.append(i['name'])
  return L

def convert3(text):
  L = []
  counter = 0
  for i in ast.literal_eval(text):
    if counter<3:
      L.append(i['name'])
      counter+= 1
    else:
      break
  return L

def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L

def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] = movies['cast'].apply(convert3)
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
```

```python
new.head()
```

```python
from gensim.models import KeyedVectors
word2vec_model = KeyedVectors.load('/content/drive/MyDrive/Colab Notebooks/word2vec-google-news-300.model')
```

```python
tags_column = new['tags']

def convert_to_word_embeddings(tags, model):
    word_embeddings = []
    for tag in tags:
        tag_words = [word for word in tag.split() if word in model.key_to_index]
        if tag_words:
            indices = [model.key_to_index[word] for word in tag_words]
            embeddings = model.vectors[indices]
            embeddings_mean = np.mean(embeddings, axis=0)  # Calculate the mean of word embeddings for the tag
        else:
            # If no word in the 'tags' is present in the word2vec model, use a default vector
            # You can choose to ignore the tags or use a zero vector
            embeddings_mean = np.zeros(model.vector_size)
        word_embeddings.append(embeddings_mean)
    return word_embeddings
```

```python
word_embeddings = convert_to_word_embeddings(tags_column, word2vec_model)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(word_embeddings)
```

```python
def recommend(movie):
    index = new[new['title_x'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title_x)
```

```python
recommend('Gandhi')
```
